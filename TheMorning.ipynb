```python
import numpy as np
import pandas as pd
import requests
from requests.exceptions import Timeout
from bs4 import BeautifulSoup
from selenium import webdriver 
from selenium.webdriver.chrome.service import Service as ChromeService 
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
import csv
```

```python
# load all the pages
driver = webdriver.Chrome()
driver.get("https://www.themorning.lk/categories/business")
def click_see_more():
    see_more_button = driver.find_element(By.XPATH, "//button[contains(text(), 'See More')]")
    see_more_button.click()
    print("Clicked 'See More' button")
    return True
driver.maximize_window()
while True:  
    try:
        click_see_more()
        time.sleep(2)
    except:
        break

```

```python
# scrape the editorial urls
elements = driver.find_elements(By.CLASS_NAME, "grid")
urls = []
for element in elements:
    anchor_elements = element.find_elements(By.TAG_NAME, "a")
    for anchor in anchor_elements:
        href = anchor.get_attribute("href")
        if href: 
            urls.append(href)
            print(href)
```

```python
print(len(urls))
```

```python
urls_1 = urls.copy()
urls_1 = list(set(urls_1))

keyword = 'articles'
filtered_urls = []
for url in urls_1:
    if keyword in url:
        filtered_urls.append(url)
        
csv_file = 'urls.csv'
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['URLs'])
    for url in filtered_urls:
        writer.writerow([url])
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import csv

# Initialize the Chrome driver with headless options
options = Options()
options.headless = True  
driver = webdriver.Chrome(options=options)

# Open the output CSV file for writing
with open('themorning_1.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['URL', 'Date', 'Title', 'Content'])

    # Open the input CSV file containing the URLs
    with open('urls.csv', 'r', newline='', encoding='utf-8') as urlfile:
        csv_reader = csv.DictReader(urlfile)
        
        # Loop through each row in the input CSV file
        for row in csv_reader:
            url = row['URLs']
            driver.get(url)
            page_source = driver.page_source
            soup = BeautifulSoup(page_source, 'html.parser')
            article_div = soup.find('article', class_='lg:col-span-2 flex flex-col items-start overflow-y-auto')

            if article_div:
                heading = article_div.find('h1', class_='font-heading font-bold text-primary-darker text-3xl sm:text-4xl lg:text-4xl mb-3').text.strip()
                date = article_div.find('div', class_='flex items-center justify-between w-full mb-6').find('p', class_='text-grey-base text-xs sm:text-base flex-shrink-0').text.strip()
                paragraphs = article_div.find_all('div', class_='first-letter:text-4xl first-letter:font-bold first-letter:mr-1 leading-7 custom-container font-articleBody')

                combined_text = ""
                for paragraph in paragraphs:
                    combined_text += paragraph.get_text().strip() + "\n"

                csv_writer.writerow([url, date, heading, combined_text])
            else:
                print("Article not found for URL:", url)

# Quit the driver
driver.quit()
```
